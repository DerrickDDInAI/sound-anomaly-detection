```python
# Importing libraries
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# load audio file ("Normal fan")
y, sr = librosa.load('/home/ms/Desktop/codit_project/00000001_-6dB_normal_fan_00.wav')
```

```python
# load audio file ("Abnormal fan")
y_abnormal, sr_abnormal = librosa.load('/home/ms/Desktop/codit_project/00000001_-6dB_Abnormal_fan_00.wav')
```

```python
# FFT normal

n_fft = 2048
ft = np.abs(librosa.stft(y[:n_fft], hop_length = n_fft+1))
plt.plot(ft);
plt.title('Spectrum')
plt.xlabel('Frequency Bin')
plt.ylabel('Amplitude');
```

```python
# FFT abnormal

n_fft = 2048
ft = np.abs(librosa.stft(y_abnormal[:n_fft], hop_length = n_fft+1))
plt.plot(ft);
plt.title('Spectrum')
plt.xlabel('Frequency Bin')
plt.ylabel('Amplitude');
```

```python
# Spectogram normal
spec = np.abs(librosa.stft(y, hop_length=512))
spec = librosa.amplitude_to_db(spec, ref=np.max)
librosa.display.specshow(spec, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram');
```

```python
# Spectogram abnormal
spec = np.abs(librosa.stft(y_abnormal, hop_length=512))
spec = librosa.amplitude_to_db(spec, ref=np.max)
librosa.display.specshow(spec, sr=sr_abnormal, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
```

```python
# Mel spectrogram normal

mel_spect = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=1024)
mel_spect = librosa.power_to_db(mel_spect, ref=np.max)

librosa.display.specshow(mel_spect, y_axis='mel', fmax=8000, x_axis='time')
plt.title('Mel Spectrogram')
plt.colorbar(format='%+2.0f dB');
```

```python
# Mel spectrogram abnormal

mel_spect_abnormal = librosa.feature.melspectrogram(y=y_abnormal, sr=sr_abnormal, n_fft=2048, hop_length=1024)
mel_spect_abnormal = librosa.power_to_db(mel_spect_abnormal, ref=np.max)

librosa.display.specshow(mel_spect_abnormal, y_axis='mel', fmax=8000, x_axis='time')
plt.title('Mel Spectrogram')
plt.colorbar(format='%+2.0f dB');
```

```python
mel_spect = librosa.feature.melspectrogram(y=y, sr=sr, n_fft=2048, hop_length=1024)
print(np.mean(mel_spect))

mel_spect_abnormal = librosa.feature.melspectrogram(y=y_abnormal, sr=sr_abnormal, n_fft=2048, hop_length=1024)
print(np.mean(mel_spect_abnormal))
```
